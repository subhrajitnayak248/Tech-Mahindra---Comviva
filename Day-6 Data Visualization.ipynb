
# Data Visualization

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

import warnings
warnings.filterwarnings("ignore")
```

```python
# load the data
data=pd.read_csv("Store.csv")
```

```python
# data['Quantity']
plt.hist(data['Quantity'],bins=30,color='green',edgecolor='black')
```

```python
sns.histplot(data['Quantity'], kde=True, color='red')
```

```python
sns.histplot(data['Quantity'], kde=True, color='red')
plt.title("Histogram for Quantity")
plt.xlabel('Quantity')
plt.ylabel('Frequency')
plt.show()
```

```python
fig=px.histogram(data['Quantity'],nbins=30,title="Histogram for Quantity")
fig.show()
```
